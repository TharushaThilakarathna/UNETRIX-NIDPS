# **1.Friday-WorkingHours-Afternoon-DDos.pcap_ISCX**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CICIDS2017/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")

print("\nLabel distribution:")
print(df[" Label"].value_counts())
```

# Clean obvious bad values (ONLY for inspection)

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

print("After cleaning:")
print(df[" Label"].value_counts())

```

# Create consistent labeling structure

```python
df[" label"] = df[" Label"].str.strip()

df[" attack_family"] = df[" label"].apply(
    lambda x: "Benign" if x == "BENIGN" else "DDoS"
)

# Keep fine-grained attack info if available
df[" attack_type"] = df[" label"].apply(
    lambda x: "Benign" if x == "BENIGN" else "DDoS"
)

df.drop(columns=[" Label"], inplace=True)
```

# Balance the dataset (1:1)

```python
attack_df = df[df[" attack_type"] == "DDoS"]
benign_df = df[df[" attack_type"] == "Benign"]

attack_sample = attack_df.sample(
    n=len(benign_df),
    random_state=42
)

balanced_df = pd.concat([attack_sample, benign_df])
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_df[" attack_type"].value_counts())
```

# Save a CLEAN SAMPLE

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CICIDS2017/DDoS_clean.csv",
    index=False
)
```

# **6.Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX**

```python
df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CICIDS2017/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv")

df[' Label'].value_counts()
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df[" label"] = df[" Label"].str.strip()

df[" label"] = df[" label"].replace({
    "Web Attack � Brute Force": "Web Attack - Brute Force",
    "Web Attack � XSS": "Web Attack - XSS",
    "Web Attack � Sql Injection": "Web Attack - SQL Injection"
})
```

```python
df[" attack_family"] = df[" label"].apply(
    lambda x: "Benign" if x == "BENIGN" else "WebAttack"
)

df[" attack_type"] = df[" label"].apply(
    lambda x: "Benign" if x == "BENIGN" else x.replace("Web Attack - ", "")
)

df.drop(columns=[" Label"], inplace=True)
```

```python
balanced_parts = []

for attack in ["Brute Force", "XSS", "SQL Injection"]:
    attack_df = df[df[" attack_type"] == attack]
    benign_df = df[df[" attack_type"] == "Benign"]

    benign_sample = benign_df.sample(
        n=len(attack_df),
        random_state=42
    )

    balanced_parts.append(pd.concat([attack_df, benign_sample]))

balanced_df = pd.concat(balanced_parts)
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_df[" attack_type"].value_counts())
```

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CICIDS2017/WebAttack_clean.csv",
    index=False
)
```

# **8.Wednesday-workingHours.pcap_ISCX**

```python
df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CICIDS2017/Wednesday-workingHours.pcap_ISCX.csv")

df[' Label'].value_counts()
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df[" label"] = df[" Label"].str.strip()
def map_attack(row):
    if row == "BENIGN":
        return "Benign", "Benign"
    elif "DoS" in row:
        return "DoS", row.replace("DoS ", "")
    elif row == "Heartbleed":
        return "Exploit", "Heartbleed"
    else:
        return "Other", row

df[[" attack_family", " attack_type"]] = df[" label"].apply(
    lambda x: pd.Series(map_attack(x))
)

df.drop(columns=[" Label"], inplace=True)
```

```python
balanced_parts = []

attack_types = ["Hulk", "GoldenEye", "slowloris", "Slowhttptest", "Heartbleed"]

for attack in attack_types:
    attack_df = df[df[" attack_type"] == attack]
    benign_df = df[df[" attack_type"] == "Benign"]

    # Special case: Heartbleed (very rare)
    if attack == "Heartbleed":
        benign_sample = benign_df.sample(
            n=len(attack_df),
            random_state=42
        )
        balanced_parts.append(pd.concat([attack_df, benign_sample]))
    else:
        benign_sample = benign_df.sample(
            n=len(attack_df),
            random_state=42
        )
        balanced_parts.append(pd.concat([attack_df, benign_sample]))

balanced_df = pd.concat(balanced_parts)
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(balanced_df[" attack_type"].value_counts())
```

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CICIDS2017/DoS_clean.csv",
    index=False
)
```
