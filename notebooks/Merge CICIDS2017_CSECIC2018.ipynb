# Load both datasets and remove spaces

```python
import pandas as pd
import numpy as np

cic = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CICIDS2017_full_clean.csv")
cse = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CSECICIDS2018_full_clean.csv")

# remove leading/trailing spaces from column names
cic.columns = cic.columns.str.strip()
cse.columns = cse.columns.str.strip()

print("CIC shape:", cic.shape)
print("CSE shape:", cse.shape)
```

# Drop unnecessary columns

```python
if "label" in cic.columns:
    cic = cic.drop(columns=["label"])
```

```python
if "Fwd Header Length.1" in cic.columns:
    cic = cic.drop(columns=["Fwd Header Length.1"])
```

```python
if "Destination Port" in cic.columns:
    cic = cic.drop(columns=["Destination Port"])
```

```python
if "Protocol" in cse.columns:
    cse = cse.drop(columns=["Protocol"])
```

# Rename the “same feature, different name” columns

```python
rename_cic = {
    "Total Length of Fwd Packets": "Fwd Packets Length Total",
    "Total Length of Bwd Packets": "Bwd Packets Length Total",
    "Min Packet Length": "Packet Length Min",
    "Max Packet Length": "Packet Length Max",
    "Average Packet Size": "Avg Packet Size",
    "Init_Win_bytes_forward": "Init Fwd Win Bytes",
    "Init_Win_bytes_backward": "Init Bwd Win Bytes",
    "act_data_pkt_fwd": "Fwd Act Data Packets",
    "min_seg_size_forward": "Fwd Seg Size Min"
}

cic = cic.rename(columns=rename_cic)
```

# Make both datasets have identical columns

```python
all_cols = sorted(list(set(cic.columns).union(set(cse.columns))))

cic = cic.reindex(columns=all_cols, fill_value=0)
cse = cse.reindex(columns=all_cols, fill_value=0)

print("Columns match:", list(cic.columns) == list(cse.columns))
print("Total columns:", len(all_cols))
```

# Merge and save

```python
merged = pd.concat([cic, cse], ignore_index=True)
print("Merged shape:", merged.shape)

merged.to_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CIC2017_CSE2018_merged_clean.csv", index=False)
print("Saved merged dataset ✅")
```
