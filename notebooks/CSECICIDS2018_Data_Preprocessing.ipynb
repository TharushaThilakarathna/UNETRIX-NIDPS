# **1. Botnet**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CSE-CICIDS2018/Botnet.csv")

df['Label'].value_counts()

```

# Clean NaN / Inf

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

# Normalize label text

```python
df['Label'] = df['Label'].str.strip()
```

# Create consistent labels

```python
df['attack_family'] = df['Label'].apply(
    lambda x: 'Benign' if x == 'Benign' else 'Botnet'
)

df['attack_type'] = df['Label'].apply(
    lambda x: 'Benign' if x == 'Benign' else 'Bot'
)

```

# Balance classes

```python
attack_df = df[df['attack_type'] == 'Bot']
benign_df = df[df['attack_type'] == 'Benign']

benign_sample = benign_df.sample(
    n=len(attack_df),
    random_state=42
)

balanced_df = pd.concat([attack_df, benign_sample])
balanced_df = balanced_df.sample(frac=1, random_state=42)
```

# Save clean dataset

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CSE-CICIDS2018/Botnet_clean.csv",
    index=False
)
```

# 5.DoS1

```python
df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CSE-CICIDS2018/DoS1.csv")

df['Label'].value_counts()
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df['Label'] = df['Label'].str.strip()

df['attack_family'] = df['Label'].apply(
    lambda x: 'Benign' if x == 'Benign' else 'DoS'
)

def map_attack_type(label):
    if label == 'Benign':
        return 'Benign'
    elif 'GoldenEye' in label:
        return 'GoldenEye'
    elif 'Slowloris' in label:
        return 'Slowloris'
    else:
        return 'DoS'

df['attack_type'] = df['Label'].apply(map_attack_type)
```

```python
attack_df = df[df['attack_family'] == 'DoS']
benign_df = df[df['attack_family'] == 'Benign']

# 1:2 ratio (safe for large datasets)
benign_sample = benign_df.sample(
    n=min(len(attack_df) * 2, len(benign_df)),
    random_state=42
)

balanced_df = pd.concat([attack_df, benign_sample])
balanced_df = balanced_df.sample(frac=1, random_state=42)
```

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CSE-CICIDS2018/DoS1_clean.csv",
    index=False
)
```

# 6.DoS2

```python
df = pd.read_csv("/content/drive/MyDrive/NIDS_Tharusha/datasets_raw/CSE-CICIDS2018/DoS2.csv")

df['Label'].value_counts()
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)
```

```python
df['Label'] = df['Label'].str.strip()
def map_attack(label):
    if label == "Benign":
        return ("Benign", "Benign")
    elif "Hulk" in label:
        return ("DoS", "Hulk")
    elif "SlowHTTPTest" in label:
        return ("DoS", "SlowHTTPTest")
    else:
        return ("DoS", "Other_DoS")

df[['attack_family', 'attack_type']] = df['Label'].apply(
    lambda x: pd.Series(map_attack(x))
)
```

```python
attack_df = df[df['attack_family'] == "DoS"]
benign_df = df[df['attack_family'] == "Benign"]

# Limit benign to max 2x attacks
max_benign = min(len(benign_df), len(attack_df) * 2)

benign_sample = benign_df.sample(
    n=max_benign,
    random_state=42
)

balanced_df = pd.concat([attack_df, benign_sample])
balanced_df = balanced_df.sample(frac=1, random_state=42)
```

```python
balanced_df.to_csv(
    "/content/drive/MyDrive/NIDS_Tharusha/datasets_clean/CSE-CICIDS2018/DoS2_clean.csv",
    index=False
)
```

```python
base_cols = dfs[0].columns

for i, df in enumerate(dfs):
    if not base_cols.equals(df.columns):
        print(f"❌ Column mismatch in file {i}")
```

```python
if "Label" in csecicids2018_full.columns:
    csecicids2018_full.drop(columns=["Label"], inplace=True)
```
